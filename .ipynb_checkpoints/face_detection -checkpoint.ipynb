```python
import cv2

# Load the Haar Cascade face detector
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Start webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

    # Draw rectangles around faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

    # Display the output
    cv2.imshow('Face Detection', frame)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Clean up
cap.release()
cv2.destroyAllWindows()
```

```python
# Paths
image_path = "face-detection-dataset/images/train/ff0b284bd67f898a.jpg"
label_path = "face-detection-dataset/labels/train/ff0b284bd67f898a.txt"

# Load and resize image
img = cv2.imread(image_path)
img = cv2.resize(img, (224, 224))
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
h_img, w_img = img.shape[:2]

# Read and parse label file
boxes = []
with open(label_path, 'r') as f:
    for line in f:
        cls, x_center, y_center, width, height = map(float, line.strip().split())
        # Convert relative coordinates to absolute
        x = int((x_center - width / 2) * w_img)
        y = int((y_center - height / 2) * h_img)
        w = int(width * w_img)
        h = int(height * h_img)
        boxes.append((x, y, w, h))

# Draw bounding boxes
for (x, y, w, h) in boxes:
    cv2.rectangle(img_rgb, (x, y), (x + w, y + h), (255, 0, 0), 1)

# Show image
plt.imshow(img_rgb)
plt.title("Image with Bounding Box")
plt.axis('off')
plt.show()
```

```python
Image_size = 224
MAX_BOXES = 5
```

```python
def load_data( image_folder, labels_folder, size = 1000 ):
    X , y = [], []
    cnt = 0
    
    for filename in os.listdir(image_folder):
        if cnt == size: break
        if filename.endswith(('.jpg', '.png')):
            
            img_path = os.path.join(image_folder, filename)
            img = cv2.imread(img_path)
            img = cv2.resize(img, (Image_size, Image_size)) 
            X.append(img/ 255.0)
    
        
            label_path= os.path.join(labels_folder, filename.replace('.jpg', '.txt') )
            boxes = []
            if os.path.exists(label_path):
                with open(label_path, 'r') as f:
                    for line in f:
                        cls, xc, yc, w, h = map(float, line.split())
                        boxes.append([xc, yc, w, h])  # Still in relative format

            if len(boxes) < MAX_BOXES:
                boxes += [[0.0, 0.0, 0.0, 0.0]] * (MAX_BOXES - len(boxes))
            else:
                boxes = boxes[:MAX_BOXES]
                
            if boxes is None:
                print("No label")
            y.append(boxes)
            
            print(img_path, " ", label_path)
            cnt += 1

    return np.array(X), np.array(y)

```

```python
train_image_folder = 'face-detection-dataset/images/train'
train_labels_folder = 'face-detection-dataset/labels/train'
X_train, y_train = load_data(train_image_folder, train_labels_folder)
```

```python
print(len(X_train), len(y_train))
```

```python
# Input image shape
Image_size = 224
MAX_BOXES = 5 
INPUT_SHAPE = (Image_size, Image_size, 3)

# Base VGG16 model
base = VGG16(weights='imagenet', include_top=False, input_shape=INPUT_SHAPE)
base.trainable = False  # Freeze base layers

# Custom head
x = base.output
x = Flatten()(x)
x = Dense(1024, activation='relu')(x)
x = Dense(MAX_BOXES * 4, activation='sigmoid')(x)  # 4 coords per box
output = Reshape((MAX_BOXES, 4))(x)

model = Model(inputs=base.input, outputs=output)
model.compile(optimizer='adam', loss='mse')

model.summary()
```

```python
model.fit(X_train, y_train, batch_size=16, epochs=20, validation_split=0.1)
```

```python
model.save("face_recog_model.keras")
```
